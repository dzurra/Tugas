## Variabel Numerikal

### SalePrice

* SalePrice merupakan variabel target yang ingin diprediksi
* Secara fisik, ini berarti harga properti yang dijual

```python
### Tulis kode Anda di sini

x = house_numerical["SalePrice"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

#### **Analisis tentang variabel `SalePrice`**

* Distribusi right-skewed, karena mean lebih besar dari median.
* Kebanyakan data berada dalam range sekitar 100,000 sampai 300,000. Apakah hal ini wajar untuk harga rumah di US?

## Pemrosesan Data Variabel Numerikal

```python
# Transformasi log-10 SalePrice

x = np.log10(house_numerical["SalePrice"])
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

```python
# Transformasi log-10 GrLivArea

x = np.log10(house_numerical["GrLivArea"])
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

```python
# Menghilangkan nilai nol pada GarageArea

x = house_numerical['GarageArea'][house_numerical['GarageArea'] !=0]
sns.distplot(x, axlabel=x.name, bins = 20)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

#### **Penjelasan dan analisis tentang pemrosesan data**

Setelah datanya ditransformasi menggunakan log-scale maka analisisnya :
* Untuk SalePrice, distribusinya right-skewed karena mean lebih besar dari median.
* Untuk GrLivArea, distribusinya left-skewed karena mean lebih kecil dari median.
* Untuk GarageArea, distribusinya right-skewed karena mean lebih besar dari median.
* Rata-rata rumah memiliki garasi seluas 500ft², dan sebanyak 81 rumah tidak memiliki garasi.

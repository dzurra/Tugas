# Data Preparation

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import statistics as stats
from google.colab import drive
drive.mount('/content/drive')

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
#Check NaN
total = house.isnull().sum().sort_values(ascending=False)
percent = (house.isnull().sum()/house.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat((total, percent), axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# Drop variabel NaN
house = house.dropna(axis=1)
house.head()
```

```python
# Memisahkan data menjadi numerikal & kategorikal
#Data Numerikal

num_cols = house._get_numeric_data().columns
house_numeric = house[num_cols]
house_numeric.head()
```

```python
#Data Kategorikal

house_categorical = house.drop(num_cols, axis=1)
house_categorical.head()
```

## Variabel Numerikal

### **SalePrice**

* SalePrice merupakan variabel target yang ingin diprediksi
* Secara fisik, ini berarti harga properti yang dijual

```python
### Tulis kode Anda di sini

fig = plt.figure()
plt.title('Grafik SalePrice', fontsize=16)
plt.xlabel('SalePrice', fontsize=14)
plt.ylabel('density', fontsize=14)

x = house_numerical["SalePrice"]
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

```python
print('number of nonzero values:', np.sum(house_numerical['SalePrice'] !=0))
print('number of zero values:', np.sum(house_numerical['SalePrice'] ==0))
```

#### **Analisis tentang variabel `SalePrice`**

* Distribusi right-skewed, karena mean lebih besar dari median.
* Tidak ada nilai 0 pada grafik ini.
* Kebanyakan data berada dalam range sekitar 100,000 sampai 300,000. Apakah hal ini wajar untuk harga rumah di US?
* Nilai selalu positif. Namun, jika dimasukkan ke model Machine Learning maka hasil prediksi bisa menjadi negatif jika salah.



### Transformasi SalePrice dengan log-scale.

* Membuat distribusi terlihat seperti normal
* Memaksa nilai harga selalu positif, bahkan ketika dimasukkan ke dalam model Machine Learning

```python
# Transformasi log-10 SalePrice

fig = plt.figure()
plt.title('Transformasi SalePrice dengan log-scale', fontsize=16)
plt.xlabel('SalePrice', fontsize=14)
plt.ylabel('density', fontsize=14)

x = np.log10(house_numerical["SalePrice"])
sns.distplot(x, axlabel=x.name)
line1 = plt.axvline(x.mean(), color="k", linestyle="--", label = "mean")
line2 = plt.axvline(np.median(x), color="r", linestyle="--", label = "median")
first_legend = plt.legend(handles=[line1, line2], loc=1)
print("Median:", x.median())
print("Mean:", x.mean())
print("Std Dev:", x.std())
plt.show()
```

# 2D Visualization

# `GrLivArea` dengan `SalePrice`

```python
### Tulis kode Anda di sini

fig = plt.figure()
plt.title('Korelasi GrLivArea dengan SalePrice', fontsize=16)
plt.xlabel('GrLivArea', fontsize=14)
plt.ylabel('SalePrice', fontsize=14)

ax = sns.scatterplot(y="SalePrice", x="GrLivArea", data = house_numerical)
plt.show()
```

### **Analisis GrLivArea dengan SalePrice**

* Variabel GrLivArea memiliki korelasi positif dengan variabel SalePrice. Namun dapat dilihat bahwa data memiliki kemungkinan anomali, yaitu titik yang berada pada ujung kanan.
* Dari scatterplot, bisa dilihat trendline dari data yang dianalisis:


```python
fig = plt.figure()
plt.title('Regplot GrLivArea dengan SalePrice', fontsize=16)
plt.xlabel('GrLivArea', fontsize=14)
plt.ylabel('SalePrice', fontsize=14)

ax = sns.regplot(y="SalePrice", x="GrLivArea", data = house_numerical)
plt.show()
```

### **Analisis Regplot**

* Korelasi antara GrLivArea dengan SalePrice ialah korelasi yang kuat. Semakin luas area tanah dan bangunannya maka secara linier harga jual pun semakin tinggi. Namun untuk nilai GrLivArea dengan nilai tertentu itu variatif.
